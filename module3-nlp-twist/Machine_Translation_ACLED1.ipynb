```python
#Retrieve pipeline of modules 
from transformers import pipeline
import pandas as pd
import csv
```

```python
#Choose English to French translation
translator = pipeline("translation_en_to_fr")
```

```python

#read csv file from github repo
acled = pd.read_csv('https://raw.githubusercontent.com/obengdouglas/DSI-2022/main/module3-nlp-twist/Data/ACLED-2022-04-01-2022-04-07-Western_Africa.csv', sep=",", nrows = 155, skip_blank_lines=True, na_filter=True)
acled.tail()
```

```python
acled.describe()
```

```python
acled["notes_fr"] = acled.notes.apply(translator)
acled.head()
```

```python
#Output saved in github repo https://github.com/obengdouglas/DSI-2022/tree/main/module3-nlp-twist/Data
acled.to_csv("acled_fr.csv", index = False)
```
